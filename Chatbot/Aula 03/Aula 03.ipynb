```python
# Importando a biblioteca
from google.colab import userdata
chat_api = userdata.get('api_chat26') # Armazenando a chave da api
```

```python
# chat_api
```

```python
from groq import Groq # Importando a biblioteca Groq
```

```python
client = Groq(api_key=chat_api) # Cria um cliente do Groq para realizar as requisições para a API

#Faz a solicitação para o modelo
completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "Você é um assistente agricultor"
        },
        {
            "role": "user",
            "content": "Qual a semente indicada para \n o solo?",
        },
    ],
    temperature = 0,
    max_tokens=1024,
    top_p=1,
    stream = False,
)

# Exibindo a resposta gerada

print(completion.choices[0].message.content)
```
